#Data visualization  
All the below charts refer to inflation rate, government debt in Thailand from 2005 - 2020.

**inflation rate per categories from 2005 - 2020**

```python
df_inflation_add.set_index('year').plot();
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

This chart shows inflation rate from 2005 - 2020 divided by category mentioned on upper right side of the chart.

**Inflation rate from 2005 - 2020**

```python
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['year']
  ys = series['inflation_rate']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df_inflation_rate_plot.sort_values('year', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('year')
_ = plt.ylabel('inflation_rate')
```

This chart shows inflation rate from 2005 - 2020 that doesn't have any specific category. Comparing to the above chart, you can see that these two charts are going to the similar direction. So we can conclude that the inflation rate has been decreased.

**Goverment debt from 2005 - 2020**

```python
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['year']
  ys = series['debt']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df_govdebt_plot.sort_values('year', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('year')
_ = plt.ylabel('debt')
```

Now let's look at the goverment debt in Thailand from 2005 - 2020. As the chart shows, the debt has been increased which is contrast to the inflation rate.
